```python
# Import the Listings.csv file (use low_memory=False and encoding="ISO-8859-1" in read_csv)
import pandas as pd

listings = pd.read_csv("C:/Users/Mattia/Desktop/PROGETTI/PYTHON/8/Airbnb Data/Listings.csv",
                       encoding="ISO-8859-1",
                       low_memory=False,
                      parse_dates=["host_since"])
```

```python
paris_listings = (
    listings
    .query("city == 'Paris'")
    .loc[:, ["host_since", "neighbourhood", "city", "accommodates", "price"]]
)

paris_listings.info()
```

```python
# OBJECTIVE 2: PREPARE FOR VISUALIZATION
```

```python
# Create A DataFrame called paris_listings_neighbourhood. Group the paris listings by neighbourhood and calculate the average price for each. 
# Sort by price in ascending order.
```

```python
paris_listings_neighbourhood = (
    paris_listings
    .groupby("neighbourhood")
    .agg({"price":"mean"})
    .sort_values("price")
)

paris_listings_neighbourhood.tail()
```

```python
# Create a DataFrame called paris_listings_accomodations. Filter your data down to the most expensive neigbourhood in Paris. 
# Group it by accomodates and calculate the average price for each. Sort by price in ascending order.
```

```python
paris_listings_accommodates = (
    paris_listings
    .query("neighbourhood =='Elysee'")
    .groupby("accommodates")
    .agg({"price":"mean","neighbourhood":"count"})
    .sort_values("price")
)

paris_listings_accommodates.head()
```

```python
# Create a DataFrame called paris_listings_over_time. Group the data by the year component of host_since. Calculate a cout of rows
# to get the number of new hosts for each year, and the average price of lisstings for each year.
```

```python
paris_listings_over_time = (
    paris_listings
    .set_index("host_since")
    .resample("YE")
    .agg({
        "neighbourhood":"count",
        "price":"mean"
    })
)

paris_listings_over_time.head()
```

```python
# Build a horizontal bar chart of average rent price by neighbourhood. Which neighbourhoods stand out?
```

```python
import seaborn as sns

(paris_listings_neighbourhood
 .plot
 .barh(
   title ="Average Listing Price by Paris Neighbourhood",
   xlabel ="Price Per Night (Euros)",
   ylabel ="Neighbourhood",
   legend=None
 )
)
sns.despine()
```

```python
(paris_listings_accommodates
 .plot
 .barh(
   title ="Average Listing Price by Accomodation Number",
   xlabel ="Price Per Night (Euros)",
   ylabel ="Accommodation Capacity",
 )
)
sns.despine()
```

```python
# Finally, build line charts of new hosts per year and average price by year. What happened to new hosts in 2015? Was average price impacted?
```

```python
paris_listings_over_time["neighbourhood"].plot(
    ylabel="New Hosts",
    title="New AirBnb Host in Paris Over Time"

)

sns.despine()
```
